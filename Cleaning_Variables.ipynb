Cleaning Variables for Analysis

5 variables were cleaned to compare work-life balance and relationship status:

**Work:**

1) Hours worked a week - "HRS1"

2) Salary - "CONRINC"

**Life:**

1) Marital Status - "MARITAL"

2) Number of kids - "CHILDS"

3) Family suffers because men overwork- "MEOVRWRK"

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Read in dataset from project1_extract_data.ipynb

```python
df = pd.read_csv('./project_gss17/gss_data_20182022.csv')
df.head()
```

```python
#1) Family suffers because men overwork - "MEOVRWRK"
meovrwrk = df['meovrwrk']
meovrwrk = meovrwrk.astype("string") # cast to string
print(meovrwrk.unique())
print(meovrwrk.dtype)
# assign cleaned data to original df
df['meovrwrk'] = meovrwrk
df['meovrwrk'] = pd.Categorical(df['meovrwrk'],categories=["strongly agree","agree","neither agree nor disagree","disagree","strongly disagree"],ordered=True)
```

```python
# 2) Marital Status - "MARITAL"
marital = df['marital']
print(marital.dtype)
# cast to string
marital = marital.astype("string")
print(marital.unique())
df['marital'] = marital
df['marital'] = pd.Categorical(df['marital'],categories=["never married","married","separated","divorced","widowed"],ordered=True)
```

```python
# 3) Number of kids - "CHILDS"
print(df['childs'].describe())
childs = df['childs']
print(childs.unique())
print(childs.dtype)
```

```python
# 4) Hours worked a week - "HRS1"
print(df['hrs1'].describe())
hrs1 = df['hrs1']
print(hrs1.unique())
print(hrs1.dtype)
```

```python
# 5) Salary - "CONRINC"
income = df['conrinc']
print(income.describe()) #Float
print(income.isna().sum()) #2,539 NA values
df.groupby('year').mean()
```
