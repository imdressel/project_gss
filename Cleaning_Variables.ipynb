**Life:**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./project_gss17/gss_data_20182022.csv')
```

```python
df.head()
```

```python
#1) Family suffers because men - "MEOVRWRK"
meovrwrk = df['meovrwrk']
meovrwrk = meovrwrk.astype("string") # cast to string
print(meovrwrk.unique())
print(meovrwrk.dtype)
```

```python
# assign cleaned data to original df
df['meovrwrk'] = meovrwrk
```

```python
df.head()
```

```python
# 2) Marital Status - "MARITAL"
marital = df['marital']
print(marital.dtype)
```

```python
# cast to string
marital = marital.astype("string")
print(marital.unique())
```

```python
df['marital'] = marital
```
