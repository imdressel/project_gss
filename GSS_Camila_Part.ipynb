```python
# CHILDS
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/project_gss17/gss_data_20182022.csv')

#Splitting into two years
df2018 = df[df['year']== 2018].copy()
df2022 = df[df['year']== 2022].copy()
```

```python
# Cleaned Variables
meovrwrk = df['meovrwrk']
meovrwrk = meovrwrk.astype("string") # cast to string
df['meovrwrk'] = meovrwrk
marital = df['marital']
marital = marital.astype("string")
df['marital'] = marital
```

# Number of Children 2018 vs. 2022

```python
df.head()
df['childs'].describe()
```

```python
#Grouped Histogram of Number of Kids 2018
sns.histplot(df2018['childs'], bins=8)
plt.xlabel('Number of Kids')
plt.title("Number of Kids 2018")
```

```python
#Grouped Histogram of Number of Kids 2022
sns.histplot(df2022['childs'], bins=8)
plt.xlabel('Number of Kids')
plt.title("Number of Kids 2022")
```

```python
#There is no considerable change in the number of children among the participants surveyed when comparing answers from 2018 and 2022.
```

```python
pd.crosstab(df['childs'],df['meovrwrk'],margins=True, normalize=True)
```

# Number of Kids vs. Marital Status

```python
#Grouped Histogram of Number of Kids vs. Marital Status
sns.histplot(data=df,x='childs',hue='marital',stat='proportion', multiple='stack', bins=9, palette='Dark2')
plt.title('Number of Kids vs. Marital Status')
plt.xlabel('Number of Kids')
plt.show()
```

```python
# Cross-tabbing Number of Kids vs. Marital Status 2018
pd.crosstab(df2018['childs'],df2018['marital'],margins=True, normalize=True)
```

```python
# Cross-tabbing Number of Kids vs. Marital Status 2022
pd.crosstab(df2022['childs'],df2022['marital'],margins=True, normalize=True)
```

CONCLUSION:
- As the number of children increases, the number of individuals not married decreases. This is likely a reflection of societal norms and various religions outlining family structures where having children outside of wedlock could be considered a "sin".
- There is no considerable change in this data from 2018 to 2022, as it can be expected that deeply rooted customs do not quickly change.

# Number of Kids vs. Men Prioritizing Work 2018 vs. 2022

```python
#Grouped Histogram of Number of Kids vs. Men Prioritizing Work 2018
df2018['meovrwrk'] = pd.Categorical(df2018['meovrwrk'],categories=["strongly agree","agree","neither agree nor disagree","disagree","strongly disagree"],ordered=True)
sns.histplot(data=df2018,x='meovrwrk',hue='childs',stat='proportion', multiple='stack', palette='Dark2')
plt.title('Number of Kids vs. Men Prioritizing Work 2018')
plt.xlabel('Men Prioritizing Work')
plt.ylabel('Number of Kids')
plt.xticks(rotation=45)
plt.show()
```

```python
#Grouped Histogram of Number of Kids vs. Men Prioritizing Work 2022
df2022['meovrwrk'] = pd.Categorical(df2022['meovrwrk'],categories=["strongly agree","agree","neither agree nor disagree","disagree","strongly disagree"],ordered=True)
sns.histplot(data=df2022,x='meovrwrk',hue='childs',stat='proportion', multiple='stack', palette='Dark2')
plt.title('Number of Kids vs. Men Prioritizing Work 2022')
plt.xlabel('Men Prioritizing Work')
plt.ylabel('Number of Kids')
plt.xticks(rotation=45)
plt.show()
```

CONCLUSION:
- More participants stated that they were unsure of whether men prioritized work after the pandemic in 2022 in comparison to before. This can be seen as a slight shift away from the larger concensus that "agreed" that family life suffers because men overwork.
- It can be infered that COVID-related work location changes are partially responsible for this shift in 2022 perception. Since many jobs transitioned to a hybrid or fully virtual schedule, individuals spent less time commuting and thus could have been perceived as working less. Individuals began to spend more time at home, and likely strengthened family relationships that were previously hindered by in-person office commitments.
