```python
#Splitting into two years
df2018 = df[df['year']== 2018].copy()
df2022 = df[df['year']== 2022].copy()
```

# Number of Kids vs. Marital Status

```python
#Grouped Histogram of Number of Kids vs. Marital Status
sns.histplot(data=df,x='childs',hue='marital',stat='proportion', multiple='stack', palette='Dark2')
plt.title('Number of Kids vs. Marital Status')
plt.xlabel('Number of Kids')
plt.ylabel('Marital Status')
plt.show()
```

# Number of Kids vs. Men Prioritizing Work 2018 vs. 2022

```python
#Grouped Histogram of Number of Kids vs. Men Prioritizing Work 2018
df2018['meovrwrk'] = pd.Categorical(df2018['meovrwrk'],categories=["strongly agree","agree","neither agree nor disagree","disagree","strongly disagree"],ordered=True)
sns.histplot(data=df2018,x='meovrwrk',hue='childs',stat='proportion', multiple='stack', palette='Dark2')
plt.title('Number of Kids vs. Men Prioritizing Work 2018')
plt.xlabel('Men Prioritizing Work')
plt.ylabel('Number of Kids')
plt.xticks(rotation=45)
plt.show()
```

```python
#Grouped Histogram of Number of Kids vs. Men Prioritizing Work 2022
df2022['meovrwrk'] = pd.Categorical(df2022['meovrwrk'],categories=["strongly agree","agree","neither agree nor disagree","disagree","strongly disagree"],ordered=True)
sns.histplot(data=df2022,x='meovrwrk',hue='childs',stat='proportion', multiple='stack', palette='Dark2')
plt.title('Number of Kids vs. Men Prioritizing Work 2018')
plt.xlabel('Men Prioritizing Work')
plt.ylabel('Number of Kids')
plt.xticks(rotation=45)
plt.show()
```
