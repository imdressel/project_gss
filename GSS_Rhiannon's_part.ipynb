# How does salary vary by marital status?

```python
#Salary by marriate type in 2018
s_by_mar18 = df18.groupby('marital')['conrinc'].mean()
s_by_mar18
```

```python
#Plot
s_by_mar18_df = pd.DataFrame(s_by_mar18)

sns.barplot(x='marital', y='conrinc', data=s_by_mar18_df, palette='viridis')
plt.title('Average Salary by Marital Status 2018')
plt.xlabel('Marital Status')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#Salary by marriage type in 2022
s_by_mar22 = df22.groupby('marital')['conrinc'].mean()
s_by_mar22
#On average - in 2022, salaries by each category of marriage type are higher than their counterparts in 2018 (besides divorced)
```

```python
s_by_mar22_df = pd.DataFrame(s_by_mar22)

sns.barplot(x='marital', y='conrinc', data=s_by_mar22_df, palette='viridis')
plt.title('Average Salary by Marital Status 2022')
plt.xlabel('Marital Status')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#Combined Plot
s_by_mar18_df['year'] = 2018
s_by_mar22_df['year'] = 2022

comb_data = pd.concat([s_by_mar18_df, s_by_mar22_df])

sns.barplot(x='marital', y='conrinc', hue='year', data=comb_data)
plt.title('Average Salary by Marital Status (2018 vs 2022)')
plt.xlabel('Marital Status')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
#On average, salary is higher in 2022 for all categories (except for slightly lower in the divorced section).
#Married people tend to have higher salaries, and never married people tend to have the lowest salary.
```

# Does salary vary by number of kids?

```python
#Salary by kids 2018
s_by_kids18 = df18.groupby('childs')['conrinc'].mean()
s_by_kids18
```

```python
#Plot
s_by_kids18_df = pd.DataFrame(s_by_kids18)

sns.barplot(x='childs', y='conrinc', data=s_by_kids18_df, palette='viridis')
plt.title('Average Salary by Number of Children (2018)')
plt.xlabel('Number of Children')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#Salary by kids 2022
s_by_kids22 = df22.groupby('childs')['conrinc'].mean()
s_by_kids22
```

```python
#Plot
s_by_kids22_df = pd.DataFrame(s_by_kids22)

sns.barplot(x='childs', y='conrinc', data=s_by_kids22_df, palette='viridis')
plt.title('Average Salary by Number of Children (2022)')
plt.xlabel('Number of Children')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.show()
```

```python
#Combined Plot
s_by_kids18_df['year'] = 2018
s_by_kids22_df['year'] = 2022

comb_data1 = pd.concat([s_by_kids18_df, s_by_kids22_df])

sns.barplot(x='childs', y='conrinc', hue='year', data=comb_data1)
plt.title('Average Salary by Number of Children (2018 vs 2022)')
plt.xlabel('Number of Children')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()
#On average, 2022 has higher salaries per number of children than 2018. It seems as if salary peaks at 2 kids, and continues to decrease after
#until there are 7 kids, which is when salary starts to increase again.
```
