```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gss = pd.read_csv('https://raw.githubusercontent.com/imdressel/project_gss17/main/gss_data_20182022.csv', low_memory=False)
print(gss.shape, '\n')
gss.head()
```

```python
income = gss['conrinc']
income.describe() #Float
#gss['conrinc'].values.tolist()
income.isna().sum() #2,539 NA values
```

```python
gss.groupby('year').mean()
#Average income 2018: $36,812.71
#Average income 2022: $42,142.82
```

```python
income_2018 = gss[gss['year'] == 2018]['conrinc']
sns.histplot(income_2018)
plt.xticks(rotation=90)
plt.show() #Graph for 2018 income
```

```python
income_2022 = gss[gss['year']== 2022]['conrinc']
sns.histplot(income_2022)
plt.xticks(rotation=90)
plt.show() #Graph for 2022 income
```
